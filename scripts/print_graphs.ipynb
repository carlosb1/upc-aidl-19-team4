```python
import csv
FIL_TRAIN = 'results/run-Jun27_01-28-09_tensorflow-1-vmsiamese1_sara_adam_normtrans_lr5e-4_v2-tag-train_av_loss.csv'
FIL_VAL = 'results/run-Jun27_01-28-09_tensorflow-1-vmsiamese1_sara_adam_normtrans_lr5e-4_v2-tag-train_val_av_loss.csv'
TEST_NAME="siamese1_sara_adam_normtrans_lr54.png"
train_fil = open(FIL_TRAIN)
reader = csv.reader(train_fil)
next(reader)
results_train = [float(row[2]) for row in reader]

val_fil = open(FIL_VAL)
reader2 = csv.reader(val_fil)
next(reader2)
results_val = [float(row[2]) for row in reader2]
print(results_train)
print("---")
print(results_val)
```

```python
import matplotlib.pyplot as plt
#plt.clf()
epochs = range(1, len(results_train) + 1)
# b is for "solid blue line"
plt.plot(epochs, results_train, 'b', label='Training loss')
# r is for "solid red line"
plt.plot(epochs, results_val, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig(TEST_NAME)
plt.show()

# epochs = range(1, len(acc_epoch_train) + 1)
# b is for "solid blue line"
# plt.plot(epochs, acc_epoch_train, 'b', label='Training accuracy')
# r is for "solid red line"
# plt.plot(epochs, acc_epoch_val, 'r', label='Validation accuracy')
# plt.title('Training and validation accuracy')
# plt.xlabel('Epochs')
#plt.ylabel('Accuracy')
# plt.legend()
# plt.savefig("train_val_acc.png")
#plt.show()
```
